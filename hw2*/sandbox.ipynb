# Градиентный спуск для функции с одной переменной

```python
def f(x):
    return x**3-5*x + 2

x_values = [x for x in np.arange(-5,5,0.1)]
f_values = [f(x) for x in np.arange(-5,5,0.1)]

plt.figure(figsize=(4,4))

plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')
plt.plot(x_values,f_values);
```

Формула для нахождения минимума методом градиентного спуска

```python
def grad_f (x):
    return 3*(x**2)-5

def gradient_descent(x_start, learning_rate, epsilon, num_iterations):
    x_curr = x_start
    
    for i in range(num_iterations):
        x_new = x_curr - learning_rate * grad_f(x_curr)
        
        if abs(x_new - x_curr) < epsilon:
            return x_curr
        
        x_curr = x_new
        
    return x_curr
```

```python
x0 = 4
eta = 0.01
epsilon = 0.001
iterations = 10000

x_res = gradient_descent(x0, eta, epsilon, iterations)

x_res
```

Нарисуем нашу точку на графике

```python
plt.figure(figsize=(4,4))

plt.axvline(x=0, c='black')
plt.axhline(y=0, c='black')
plt.plot(x_values, f_values)

plt.scatter(x_res, f(x_res),marker='o', c='red');
```

# Градиентный спуск для функции с несколькими переменными

```python
RANDOM_STATE = 42
np.random.RandomState(RANDOM_STATE);
```

```python
df = pd.DataFrame({'x1':[10,5,7]
                  ,'x2':[40,36,30]
                  ,'y':[100,73,71]})
df
```

```python
def prediction(w, x1, x2):
    return w[0] + w[1] * x1 + w[2] * x2
```

```python
def grad_MSE(w,x1,x2,y):
    dw0 = (2/3) * (w[0] + w[1]*x1 + w[2]*x2 - y)
    dw1 = (2/3) * x1 * (w[0] + w[1]*x1 + w[2]*x2 - y)
    dw2 = (2/3) * x2 * (w[0] + w[1]*x1 + w[2]*x2 - y)
    
    return np.array([dw0, dw1, dw2])
```

```python
grad_MSE((1,1,1), 10, 40, 100)
```

```python
def gradient_descent(data, w_start, learning_rate, num_iterations):
    w_curr = w_start
    
    for i in range(num_iterations):
        GradMSE = np.zeros(3)
        for _, row in data.iterrows():
            x1,x2,y = row
            GradMSE += grad_MSE(w_curr, x1,x2, y)
            
        w_new = w_curr - learning_rate * GradMSE
        
        w_curr = w_new
        
    return w_curr
```

```python
gradient_descent(df, np.random.rand(3), 0.0001, 5000)
```

# Векторная форма градиентного спуска

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random

RANDOM_STATE = 42
np.random.RandomState(RANDOM_STATE);
```

```python
df = pd.DataFrame({'x1':[10,5,7]
                  ,'x2':[40,36,30]
                  ,'y':[100,73,71]})

df
```

Функция для вычисления MSE

```python
def compute_cost(X,y,theta):
    m = len(y)
    
    cost = (1./m) * (np.linalg.norm(X @ theta - y) ** 2)
    return cost
```

```python
def gradient_descent(X,y,learning_rate,iterations):
    X = np.hstack((np.ones((X.shape[0], 1)), X)) # Добавляем к X единички
    params = np.random.rand(X.shape[1])
    
    m = X.shape[0]
    
    cost_track = np.zeros((iterations,1))
    
    for i in range(iterations):
        params = params - 2./m * learning_rate * (X.T @ ((X @ params) - y))
        cost_track[i] = compute_cost(X,y,params)
        
    return cost_track, params
```

```python
def predict(X, params):
    X = np.hstack((np.ones((X.shape[0], 1)),X))
    return X @ params
```
